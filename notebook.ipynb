```python
import yfinance as yf
import pandas as pd

df = yf.download("AAPL", period="1mo", interval="1d")
df = df.reset_index()  
print(df.columns)

print(df.head())
df.columns = df.columns.get_level_values("Price")
df.columns.name = None
print(df.head())
print(df.columns)
```

```python
import warnings 

def compute_daily_returns(df):
    df = df.copy()
    df["Return"] = df["Close"].pct_change()
    return df
df=compute_daily_returns(df)
def summary_stats(df):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")

    first = df["Close"].iloc[0]
    last = df["Close"].iloc[-1]

    full_return = (last / first) - 1
    stats =  {
        "Mean return": float(df["Return"].mean()),
        "Volatility": float(df["Return"].std()),
        "Min price": float(df["Close"].min()),
        "Max price": float(df["Close"].max()),
        "Full return": float(full_return)
    }
    return pd.DataFrame(
        [(k, v) for k, v in stats.items()],
        columns=["Metric", "Value"]
    )
summary_stats(df)
```
